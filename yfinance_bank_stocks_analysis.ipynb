
### Logarithmic Returns:
Logarithmic (or log) returns are often preferred in financial modeling because they are time-additive, meaning that the log return over multiple periods is simply the sum of the log returns over each sub-period. This property simplifies many calculations and is more appropriate for continuous-time processes.

**Formula:**
The formula for logarithmic daily return (r_t) on day 't' is:

r_t = ln(P_t / P_{t-1})

Where:
*   `P_t` is the closing price on day 't'.
*   `P_{t-1}` is the closing price on the previous day (t-1).
*   `ln` denotes the natural logarithm.

Logarithmic returns are symmetric, so a 10% gain followed by a 10% loss will bring the price back to its original value. This is not true for simple returns, where (1+0.10)(1-0.10) = 0.99, meaning a 1% loss.

```python
import numpy as np

log_returns = np.log(close_prices_data / close_prices_data.shift(1))

print("First 5 rows of logarithmic returns:")
print(log_returns.head())

print("\nLast 5 rows of logarithmic returns:")
print(log_returns.tail())
```


### Cumulative Returns:
Cumulative returns measure the total gain or loss of an investment over a period of time, compounding all intermediate returns. They provide a comprehensive view of an investment's performance from its starting point to its end point, making them essential for evaluating long-term investment strategies.

**Formula:**
The formula for cumulative return (CR_n) over 'n' periods is:

CR_n = (1 + R_1) * (1 + R_2) * ... * (1 + R_n) - 1

Where:
* `R_i` represents the simple daily return for period 'i'.

```python
cumulative_returns = pd.DataFrame()

for column in daily_percentage_change.columns:
    # Convert percentage change to decimal and add 1
    daily_returns = 1 + (daily_percentage_change[column] / 100)

    # Calculate cumulative product
    cumulative_product = daily_returns.cumprod()

    # Subtract 1 and multiply by 100 to get cumulative return as a percentage
    new_column_name = column.replace("('Close', ", "Cumulative_Return_").replace(")", "")
    cumulative_returns[new_column_name] = (cumulative_product - 1) * 100

print("First 5 rows of cumulative returns:")
print(cumulative_returns.head())

print("\nLast 5 rows of cumulative returns:")
print(cumulative_returns.tail())
```


### Absolute Monthly Change:
Absolute Monthly Change refers to the raw difference in a stock's closing price from the end of one month to the end of the next. It quantifies the exact monetary gain or loss over a monthly period, without considering the initial price.

**Formula:**
The formula for Absolute Monthly Change (\( \Delta P \)) is:

\( \Delta P = P_{\text{end of month}} - P_{\text{previous end of month}} \)

Where:
*   `P_end of month` is the closing price at the end of the current month.
*   `P_previous end of month` is the closing price at the end of the previous month.

This calculation tells us the exact amount by which the price has changed from one month to the next. A positive value indicates an absolute gain, while a negative value indicates an absolute loss.

The `monthly_differences` DataFrame, calculated earlier, already contains these values, representing the absolute change between consecutive monthly closing prices for each stock.


### Daily Volatility:
Daily volatility is a measure of the dispersion of daily returns for a given stock or asset. It quantifies how much the price of an asset fluctuates over a single trading day. High volatility indicates that the asset's price can change dramatically over a short period, while low volatility suggests more stable price movements.

**Significance:**
Daily volatility is crucial for risk management, option pricing, and portfolio optimization. Traders and investors use it to gauge the risk associated with an investment. Higher volatility typically implies higher risk but also potentially higher returns. It's an essential component in calculating risk-adjusted returns and understanding market uncertainty.

**Formula:**
Daily volatility is typically measured as the standard deviation of an asset's daily returns. While the daily returns themselves (simple or logarithmic) represent the change, the standard deviation of these returns over a period (e.g., a year) gives us the volatility.

Specifically, if we have a series of daily returns $R_1, R_2, ..., R_n$, the daily volatility ($\sigma_{\text{daily}}$) can be calculated as:

$\sigma_{\text{daily}} = \text{Standard Deviation}(R_t)$

To annualize daily volatility, it is commonly scaled by the square root of the number of trading days in a year (e.g., 252 for stocks):

$\sigma_{\text{annual}} = \sigma_{\text{daily}} \times \sqrt{\text{Number of Trading Days}}$

Where:
*   $R_t$ represents the daily returns (either simple or logarithmic).
*   $\text{Standard Deviation}(R_t)$ is the standard deviation of the daily returns.

```python
daily_volatility = daily_percentage_change.std(axis=0, skipna=True)

print("Daily Volatility (Standard Deviation of Daily Percentage Change):\n")
print(daily_volatility)
```

### Annualized Volatility:
Annualized volatility is a measure of the expected dispersion of returns of a financial instrument over a year. It's an important metric for investors to assess the risk of an asset over a longer period, allowing for comparison between assets with different trading frequencies or data intervals. Annualizing volatility helps standardize risk measurement.

**Formula:**
If you have daily volatility (standard deviation of daily returns), you can annualize it using the following formula:

$ \sigma_{\text{annual}} = \sigma_{\text{daily}} \times \sqrt{N} $

Where:
*   $ \sigma_{\text{annual}} $ is the annualized volatility.
*   $ \sigma_{\text{daily}} $ is the daily volatility (standard deviation of daily returns).
*   $ N $ is the number of trading days in a year (commonly 252 for stocks, or 365 for assets traded every day).

```python
annualized_volatility = daily_volatility * np.sqrt(252)

print("Annualized Volatility (Standard Deviation of Daily Percentage Change):")
print(annualized_volatility)
```


### Sharpe Ratio:
The Sharpe Ratio is a measure of risk-adjusted return of an investment. It is used to help investors understand the return of an investment compared to its risk. The ratio is the average return earned in excess of the risk-free rate per unit of volatility or total risk. A higher Sharpe Ratio indicates better risk-adjusted performance.

**Significance:**
The Sharpe Ratio is widely used by fund managers, analysts, and investors to evaluate the performance of portfolios. It helps in making investment decisions by providing a standardized measure to compare different investments, taking into account the risk involved. It's particularly useful when comparing two investments with similar returns but different levels of risk.

**Formula:**
The formula for the Sharpe Ratio is:

$ \text{Sharpe Ratio} = \frac{\text{Average Return} - \text{Risk-Free Rate}}{\text{Annualized Volatility}} $

**Components of the Formula:**
*   **Average Return:** This is the average rate of return on the investment over a specified period. In our context, this will typically be the `annual_returns` for each stock.
*   **Risk-Free Rate:** This is the return on an investment with zero risk, such as a short-term government bond. It represents the return an investor could expect without taking on any investment risk. For our analysis, we will need to assume or define a risk-free rate.
*   **Annualized Volatility:** This is the standard deviation of the investment's returns, scaled to an annual period. It measures the total risk or fluctuations in the investment's value. We have already calculated this as `annualized_volatility` for each stock.

```python
risk_free_rate = 0
sharpe_ratio = (annual_returns - risk_free_rate) / annualized_volatility

print("Sharpe Ratio for each stock (Risk-Free Rate = 0):")
print(sharpe_ratio)
```


### Maximum Drawdown (MDD):
Maximum Drawdown (MDD) is a key risk metric that represents the largest percentage drop from a peak to a trough in the value of an investment (or portfolio) before a new peak is achieved. It measures the downside risk over a specified period and indicates the worst possible loss an investor would have incurred if they had bought at the peak and sold at the bottom.

**Significance:**
MDD is crucial for risk management and performance evaluation. It helps investors understand the potential worst-case scenario for an investment, which is vital for setting risk tolerance levels and making informed decisions. A lower MDD is generally preferred, as it indicates better capital preservation during market downturns.

**Formula:**
The formula for Maximum Drawdown is:

$ \text{MDD} = \frac{\text{Trough Value} - \text{Peak Value}}{\text{Peak Value}} $

Where:
*   **Peak Value:** The highest point (value) of the investment before a decline.
*   **Trough Value:** The lowest point (value) reached before a new peak is established.

Alternatively, it can be calculated as:

$ \text{MDD} = \max_{\tau \in [t, T]} \left( \frac{P_t - P_{\tau}}{P_t} \right) $

Where:
*   $ P_t $ is the asset value at time $ t $.
*   $ P_{\tau} $ is the asset value at time $ \tau $, where $ \tau > t $.
*   $ \text{Max} $ denotes taking the maximum over all possible periods from $ t $ to $ T $.

```python
running_max = cumulative_returns.cummax()
drawdown = (running_max - cumulative_returns) / running_max

maximum_drawdown = drawdown.max() * 100 # Express as percentage

print("Maximum Drawdown for each stock:")
print(maximum_drawdown)
```


### Correlation Analysis:
Correlation analysis is a statistical method used to evaluate the strength and direction of a linear relationship between two or more financial variables, such as stock returns. In portfolio management, understanding the correlation between different assets is crucial for diversification.

**Significance:**

*   **Diversification**: Low or negative correlation between assets allows for portfolio diversification, meaning that when one asset performs poorly, another might perform well, thereby reducing overall portfolio risk.
*   **Risk Management**: Highly correlated assets tend to move in the same direction, which can increase portfolio risk during market downturns. Identifying and managing these correlations is essential for robust risk management.
*   **Strategy Development**: Correlation insights can inform trading strategies, such as pairs trading, where two highly correlated stocks are traded based on their relative performance.

**Formula:**
The Pearson correlation coefficient (often denoted as \( \rho \) or \( r \)) between two variables X and Y is given by:

$$ \rho_{X,Y} = \frac{\text{cov}(X, Y)}{\sigma_X \sigma_Y} $$

Where:
*   `cov(X, Y)` is the covariance between variables X and Y.
*   `\( \sigma_X \)` is the standard deviation of X.
*   `\( \sigma_Y \)` is the standard deviation of Y.

**Interpretation:**
*   The correlation coefficient ranges from -1 to +1.

## Summary:

### Data Analysis Key Findings

The analysis involved introducing and calculating several key financial metrics for a set of banking stocks.

*   **Simple Daily Returns**: This metric measures the percentage change in a stock's price from one day to the next, calculated as \(R_t = (P_t - P_{t-1}) / P_{t-1}\). The `daily_percentage_change` DataFrame was confirmed to hold these values, expressed as percentages (e.g., -1.22% or 0.92%).
*   **Logarithmic Returns**: Preferred in financial modeling for their time-additive property and symmetry, calculated as \(r_t = \ln(P_t / P_{t-1})\). These were successfully computed and stored in the `log_returns` DataFrame.
*   **Cumulative Returns**: This metric tracks the total gain or loss of an investment over time by compounding intermediate returns, calculated as \(CR_n = (1 + R_1) * (1 + R_2) * ... * (1 + R_n) - 1\). The cumulative returns for each stock were calculated and visualized, showing their performance trajectories.
*   **Monthly Resampling**: Explained as a process to aggregate daily data into monthly frequencies for long-term trend analysis, typically by taking the last closing price of each month.
*   **Monthly Percentage Returns**: Measures the percentage change in price from the last trading day of one month to the last trading day of the next month, using the formula \(R_{monthly} = (P_{end\_month} - P_{start\_month}) / P_{start\_month}\).
*   **Absolute Monthly Change**: Quantifies the raw monetary difference in a stock's closing price between consecutive month-ends, calculated as \(\Delta P = P_{\text{end of month}} - P_{\text{previous end of month}}\).
*   **Daily Volatility**: Measures the dispersion of daily returns, typically as the standard deviation of daily returns. For the analyzed banking stocks, values ranged from approximately 1.39% (HDFCBANK.NS) to 2.65% (PNB.NS).
*   **Annualized Volatility**: Scales daily volatility to an annual period for standardized risk measurement, using the formula \(\sigma_{\text{annual}} = \sigma_{\text{daily}} \times \sqrt{252}\). Annualized volatility for the stocks ranged from approximately 21.99% (HDFCBANK.NS) to 42.04% (PNB.NS).
*   **Sharpe Ratio**: A key risk-adjusted return metric, calculated as \( (\text{Average Return} - \text{Risk-Free Rate}) / \text{Annualized Volatility} \). With a risk-free rate of 0, HDFCBANK.NS exhibited the highest Sharpe Ratio at 0.7086, indicating superior risk-adjusted performance among the selected stocks.
*   **Maximum Drawdown (MDD)**: Represents the largest percentage drop from a peak to a trough, indicating potential worst-case loss. Calculated MDD values were substantial, ranging from 90.49% (AXISBANK.NS) to 1063.48% (KOTAKBANK.NS), though extreme values may warrant further data verification.
*   **Correlation Analysis**: Examines the strength and direction of linear relationships between stock returns. A heatmap visualization showed that most banking stocks exhibited positive correlations (e.g., ICICIBANK.NS and HDFCBANK.NS with a 0.52 correlation), which is expected within the same sector, suggesting similar reactions to market events.
